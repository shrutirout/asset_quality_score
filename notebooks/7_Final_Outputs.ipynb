```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Define all results
results = {
    "Logistic Regression":     [0.9074, 0.7628, 0.3425, 0.4727, 0.8903],
    "Decision Tree":           [0.8856, 0.6672, 0.1134, 0.1938, 0.7783],
    "Random Forest":           [0.8981, 0.9013, 0.1793, 0.2991, 0.8870],
    "XGBoost":                 [0.9415, 0.9388, 0.5534, 0.6964, 0.9551],
    "Logistic Regression (SMOTE)": [0.8080, 0.3674, 0.8087, 0.5052, 0.8925],
    "SVM":                     [0.9016, 0.7873, 0.2587, 0.3895, 0.8879],
    "k-NN":                    [0.8755, 0.4513, 0.1256, 0.1965, 0.6730],
    "TensorFlow":              [0.9610, 0.9379, 0.7261, 0.8185, 0.9657],
    "PyTorch":                 [0.9530, 0.9006, 0.6884, 0.7804, 0.9561],
}

metrics_df = pd.DataFrame.from_dict(
    results, orient='index',
    columns=["Accuracy", "Precision", "Recall", "F1 Score", "AUC"]
).sort_values("AUC", ascending=False)

print("\nðŸ“Š Final Model Comparison Table:")
print(metrics_df.round(4))

# Step 2: Plotting
plt.figure(figsize=(14, 7))
metrics_df[["F1 Score", "AUC","Accuracy"]].plot(kind='bar', figsize=(14, 6), colormap='viridis')
plt.title("Model Comparison: F1 Score, AUC And Accuracy")
plt.ylabel("Score")
plt.xticks(rotation=45, ha='right')
plt.grid(True, axis='y')
plt.tight_layout()
plt.legend(loc='lower right')
plt.show()

# Step 3: Identify best model by AUC and F1
best_by_auc = metrics_df['AUC'].idxmax()
best_by_f1 = metrics_df['F1 Score'].idxmax()

print(f"\n Best model by AUC: {best_by_auc}")
print(f" Best model by F1 Score: {best_by_f1}")
```
